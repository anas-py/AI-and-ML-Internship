```python
import pandas as pd

# Load dataset
df = pd.read_csv('Titanic-Dataset.csv')
```

```python
df
```

```python

# Display basic info and first few rows
df.info()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Set plot style
sns.set(style="whitegrid")

# 1 Summary statistics
summary_stats = df.describe(include='all')
```

```python
summary_stats
```

```python
# 2 Histograms for numeric features
numeric_cols = ['Age', 'Fare', 'SibSp', 'Parch']
fig, axes = plt.subplots(2, 2, figsize=(12, 8))
axes = axes.flatten()
for i, col in enumerate(numeric_cols):
    sns.histplot(data=df, x=col, kde=True, ax=axes[i], color='skyblue')
    axes[i].set_title(f'Histogram of {col}')
plt.tight_layout()
plt.show()
```

```python
# 3 Boxplots for numeric features by Survived
fig, axes = plt.subplots(2, 2, figsize=(12, 8))
axes = axes.flatten()
for i, col in enumerate(numeric_cols):
    sns.boxplot(data=df, x='Survived', y=col, ax=axes[i])
    axes[i].set_title(f'Boxplot of {col} by Survival')
plt.tight_layout()
plt.show()
```

```python
# 4 Correlation heatmap
corr_matrix = df[['Survived', 'Pclass', 'Age', 'SibSp', 'Parch', 'Fare']].corr()
plt.figure(figsize=(8, 6))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', fmt=".2f")
plt.title("Correlation Heatmap")
plt.show()
```
